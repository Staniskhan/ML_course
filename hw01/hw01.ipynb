```python
import sys 
import numpy as np
import pandas as pd
import sklearn

print("python version:", sys.version)
print("numpy version:", np.__version__)
print("pandas version:", pd.__version__)
print("sklearn version:", sklearn.__version__)
```

```python
np.random.seed(100)
matr = np.random.random((100, 5))

print("shape:", matr.shape)
print("dtype:", matr.dtype)
print("first 5 rows:\n", matr[0:5])
```

```python
print("avg:", np.mean(matr))
print("std:", np.std(matr))
print("min:", np.min(matr))
print("max:", np.max(matr))
print("avg (axis=0):", np.mean(matr, axis=0))
print("std (axis=0):", np.std(matr, axis=0))
print("min (axis=0):", np.min(matr, axis=0))
print("max (axis=0):", np.max(matr, axis=0))
```

```python
w = np.random.random((5))
y = matr @ w
print("y.shape:", y.shape)
```

```python
w_true = np.array([1, 2, 3, 4, 5])
noise = np.random.random((100)) * 0.1
y_true = matr @ w_true + noise

w_guess = w_true + np.random.random((5)) * 0.5 - 0.25
y_pred = matr @ w_guess

mse = np.mean((y_true - y_pred) ** 2)
print("MSE:", mse)
mae = np.mean(np.abs(y_true - y_pred))
print("MAE:", mae)
rmse = np.sqrt(mse)
print("RMSE:", rmse)
r2 = 1 - np.sum((y_true - y_pred) ** 2) / np.sum((y_true - np.mean(y_true)) ** 2)
print("R^2:", r2)
```
